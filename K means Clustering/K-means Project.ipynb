## K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
#We will set to 2 the number of clusters ...
kmeans = KMeans(n_clusters=2)
```

```python
#We fit the model to d without the Private column
#d.drop(d.index[776],inplace=True)
x = a.drop('Private',axis=1)
kmeans.fit(x)
```

```python
#cluster center vectors
kmeans.cluster_centers_
```
